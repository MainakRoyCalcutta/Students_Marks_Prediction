# Visualization of Data

## import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("test_scores.csv")
```

## Data Visualization

```python
public = dataset[dataset.school_type == "Public"]
non_public = dataset[dataset.school_type != "Public"]
```

```python
#Public School Student vs Pretest Marks
plt.figure(figsize=(15,10))
plt.plot(public.n_student, public.pretest, 'o')
```

```python
#Non-Public School Student vs Pretest Marks
plt.figure(figsize=(15,10))
plt.plot(non_public.n_student, non_public.pretest, 'o')
```

```python
#Student Gender vs Pretest Marks
plt.figure(figsize=(15,10))
plt.plot(dataset.gender, dataset.pretest, 'o')
```

```python
#School Type vs Pretest Marks
plt.figure(figsize=(15,10))
plt.plot(dataset.school_type, dataset.pretest, 'o')
```

```python
#Teaching method vs Pretest Marks
plt.figure(figsize=(15,10))
plt.plot(dataset.teaching_method, dataset.pretest, 'o')
```

```python
#School Setting vs Pretest Marks
plt.figure(figsize=(15,10))
plt.plot(dataset.school_setting, dataset.pretest, 'o')
```
